# IPL 2018 Data Analysis on _Matches_

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as s
import numpy as np
```

```python
df=pd.read_csv("Desktop\DATA SETS\IPL\matches.csv")
df
```

```python
dff=df[df["season"]==2018]
dff
```

# Teams and their *toss winning* analysis :

```python
s.countplot(x="toss_winner",data=dff)
plt.xlabel('Teams')
plt.ylabel('No of matches')
plt.xticks(rotation='vertical')
plt.show()
```

# Choices of Toss winners:

```python
s.countplot(x="toss_winner",hue='toss_decision',data=dff)
plt.xlabel('Teams')
plt.ylabel('No of matches')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#adding a column for checking whether the toss winner is match winner or not?
```

```python
dff['tosswinner_matchwinner']=np.where(dff['toss_winner']==dff['winner'],'win','lose')
dff.head()
```

## Did _Tosswinners_ won the match?

```python
s.countplot(x="tosswinner_matchwinner",data=dff)
plt.xlabel('Do toss-winners win the match?')
plt.ylabel('No of matches')
plt.show()
```

## Teamwise Tosswinning-MatchWinning Analysis:

```python
plt.figure(figsize=(10,7))
s.countplot(x="tosswinner_matchwinner",hue='toss_winner',data=dff)
plt.xlabel('Did a team win after winning the toss')
plt.ylabel('No of matches')
plt.show()
```

```python
#adding one more column for checking chasing and defending
```

```python
dff['target']=np.where((dff['win_by_runs']==0)&(dff['win_by_wickets']!=0),'chased','defended')
dff.head()
```

## Data of Chased and Defended Targets :

```python
plt.figure(figsize=(7,7))
s.countplot("target",data=dff)
plt.ylabel('No of matches')
```

## Let's see,this year  Which Team is Good at Chasing and which is good at defending ?

```python
plt.figure(figsize=(7,7))
s.countplot(x='winner',hue='target',data=dff)
plt.xticks(rotation='vertical')
plt.ylabel('No.of.Matches')
```

### _It is clear that __CSK__ is good at chasing and __SRH__ is good at defending !!!_
